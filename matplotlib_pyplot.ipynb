# Plotting with Pyplot API

## Using `pyplot` only

`pyplot` contains a simpler interface for creating visualizations that allow us to plot the data without explicitly configuring the Figure and Axes objects. It is meant for people with less programming experience or that do not need full control over the visuals.

```python
# create a figure
plt.figure(figsize=(12,7))
# plot
plt.plot([2, 4, 6, 2, 10], color='green', marker='o', linestyle='-.', linewidth=1.5, markersize=8)
# set ticks
plt.xticks(ticks=np.arange(5))
plt.show();
```

```python
# create a figure and add title to the current figure
fig = plt.figure(figsize=(12,7))
fig.suptitle("Performance Analysis", fontsize=(16), fontweight="bold")
# plots
plt.plot([2.2, -0.5, 3.8, 2., 1.2], color='blue', marker='o', linestyle='-.', linewidth=1.5, markersize=8, label="Return % week 1")
plt.plot([-1.3, 1.1, 0.2, 5.2, 3.2], color='orange', marker='o', linestyle='-.', linewidth=1.5, markersize=8, label="Return % week 2")
# set ticks and add labels
plt.xticks(ticks=np.arange(5), labels=["Mon", "Tue", "Wed", "Thu", "Fri"])
# add axis label
plt.xlabel("Day")
plt.ylabel("Return %")
# add title to current axes
plt.title("Daily returns", fontsize=(12))
# add 0 line
plt.axhline(y=0, color="black", linewidth=1)
# add hoz line max and min
plt.axhline(y=5.2, color="green", linewidth=3)
plt.text(3., 5.3, "Max daily return", color="green", fontweight="bold")
plt.axhline(y=-1.3, color="red", linewidth=3)
plt.text(0.15, -1.6, "Min daily return", color="red", fontweight="bold")
# add max and min daily returns
plt.text(2.75, 5.3, "5.3 %",fontweight="bold", color="green")
plt.text(-0.1, -1.6, "-1.4 %",fontweight="bold", color="red")
# set y limits
plt.ylim((-3, 6))
# add grid
plt.grid(axis="y", which="major", linestyle="--", color="lightgray")
# add legend
plt.legend()
plt.show();
```
