```python
import matplotlib.pyplot as plt

# Set the backend as the screen output
%matplotlib inline
```

```python
# Get the location of matplotlibrc file
import matplotlib
matplotlib.matplotlib_fname()
```

```python
print(matplotlib.rcParams)
```

```python
# plot something
plt.plot([1., 3., 4.]);
# Add labels and title
plt.title("Toy Plot") #Prints the title on top of graph
plt.xlabel("x-axis") # Prints X axis label as "X-axis"
plt.ylabel("y-axis") # Prints Y axis label as "Y-axis
```

```python
# change some defaults
matplotlib.rc('lines', linewidth=2, linestyle='--', marker='+')
matplotlib.rcParams['lines.markersize'] = 15
matplotlib.rcParams['font.size'] = '15.0'
matplotlib.rcParams['axes.edgecolor'] = "blue"
```

```python
# plot something again
plt.plot([1., 3., 4.]);
# Add labels and title
plt.title("Toy Plot") #Prints the title on top of graph
plt.xlabel("x-axis") # Prints X axis label as "X-axis"
plt.ylabel("y-axis") # Prints Y axis label as "Y-axis
```

```python
# To restore all default parameters
matplotlib.rcdefaults() 
```
