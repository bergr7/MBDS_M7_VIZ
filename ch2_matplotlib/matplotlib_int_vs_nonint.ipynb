```python
import matplotlib as mpl
import matplotlib.pyplot as plt

# Set the backend as the screen output
%matplotlib inline
```

# Modo interactivo

```python
# default mode is interactive
mpl.is_interactive()
```

In the interactive mode, the graph display gets updated after each statement.

```python
# plot something
plt.plot([1., 3., 4.]);
# Add labels and title
plt.title("Interactive Plot") #Prints the title on top of graph
plt.xlabel("X-axis") # Prints X axis label as "X-axis"
plt.ylabel("Y-axis") # Prints Y axis label as "Y-axis"
```

# Modo No Interactivo

```python
# Set no interactive mode and double check
plt.ioff()
mpl.is_interactive()
```

In Jupyter Notebooks you will get the figure automatically but if you use the python shell, you need need to ask specifically to plot if you are working on non-interactive mode.

```python
# plot something
plt.plot([1.5, 3.0])
plt.title("Non-interactive Plot") #Prints the title on top of graph
```

Ejecutar la celda anterior no produce ninguna salida. Necesitamos pedir a maplotlib específicamente que ejecute cada declaración.

```python
plt.show()
```

Una vez que pedimos que se dibuje el gráfico, ya no podemos seguir añadiendo elementos o haciendo cambios.
