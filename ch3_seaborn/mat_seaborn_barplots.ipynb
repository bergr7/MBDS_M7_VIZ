```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

rs = np.random.RandomState(8)
```

En primer lugar vamos a fijar el tema que queremos para la visualización. Esto lo podemos hacer con el método `.set_theme()`

```python
sns.set_theme(style="white", context="talk")
```

Este tema se aplicará a todas las visualizaciones que creemos con seaborn y matplotlib.

Vamos a crear el obejto figure con matplotlib. En este caso, vamos a tener 3 filas y una columna en la figura. El tamaño de la figura va a ser de 7 in x 5 in y las 3 figuras comparten el eje x (sharex=True)

```python
# demostración de las figuras en blanco (modo interactivo)
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(7, 5), sharex=True)
```

Vamos a generar datos sintéticos para este ejemplo.

```python
x = np.array(list("ABCDEFGHIJ"))
y1 = np.arange(1, 11)
y2 = y1 - 5.5
y3 = rs.choice(y1, len(y1), replace=False)
```

Para crear las visualizaciones, vamos a utilizar el método `sns.barplot()` de seaborn para crear la figura, el método `.ahline()` del objeto axes de matplotlib para una línea horizontal y el método `.set_ylabel()` del objeto axes de matplotlib para fijar la etiqueta del eje y.

```python
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(7, 5), sharex=True)

# figura 1
sns.barplot(x=x, y=y1, palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Sequential")

# figura 2
sns.barplot(x=x, y=y2, palette="vlag", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Diverging")

# figura 3
sns.barplot(x=x, y=y3, palette="deep", ax=ax3)
ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("Qualitative")

# quitar spines
sns.despine(bottom=True)

# quitar los yticks de los axes de la figura
plt.setp(f.axes, yticks=[])

# ajustar el padding
plt.tight_layout(h_pad=2)
```
